## Section 7: Missing Value Analysis

Calculate percentage missing:

```python
missing = (
    df.isnull()
      .mean()
      .sort_values(ascending=False)
      * 100
)

missing.head(20)
```

Create summary dataframe:

```python
missing_df = pd.DataFrame({
    "Feature": missing.index,
    "Missing_Percentage": missing.values
})

missing_df.head(20)
```

Visualize top missing features:

```python
plt.figure(figsize=(12,6))

missing_df.head(20).plot(
    x="Feature",
    y="Missing_Percentage",
    kind="bar"
)

plt.title("Top 20 Missing Features")
plt.ylabel("% Missing")
plt.show()
```

### Missing Value Observations

Several variables contain substantial missing values.

Missingness itself may contain predictive information and should be investigated before removal.

## Section 11: Categorical Feature Analysis

Important categorical variables:

```python
important_cat = [
    "CODE_GENDER",
    "NAME_INCOME_TYPE",
    "NAME_EDUCATION_TYPE",
    "NAME_FAMILY_STATUS"
]
```

Default rates by category:

```python
for col in important_cat:
    print("\n")
    print(col)
    display(
        pd.crosstab(
            df[col],
            df["TARGET"],
            normalize="index"
        )
    )
```

Visualization:

```python
for col in important_cat:
    plt.figure(figsize=(10,4))
    sns.barplot(x=df[col], y=df["TARGET"])
    plt.xticks(rotation=45)
    plt.title(f"Default Rate by {col}")
    plt.show()
```

## Section 12: Correlation Analysis

Correlation with target:

```python
corr_target = (
    df.corr(numeric_only=True)["TARGET"]
      .sort_values(ascending=False)
)

corr_target.head(20)
```

Strongest positive correlations:

```python
corr_target.head(15)
```

Strongest negative correlations:

```python
corr_target.tail(15)
```

## Section 13: Correlation Heatmap

Use selected features only:

```python
important_features = [
    "TARGET",
    "AMT_INCOME_TOTAL",
    "AMT_CREDIT",
    "AMT_ANNUITY",
    "DAYS_BIRTH",
    "DAYS_EMPLOYED",
    "EXT_SOURCE_1",
    "EXT_SOURCE_2",
    "EXT_SOURCE_3"
]
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df[important_features].corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()
```

## Section 14: Top Features Correlated with TARGET

```python
corr_abs = (
    df.corr(numeric_only=True)["TARGET"]
      .drop("TARGET")
      .abs()
      .sort_values(ascending=False)
)

top_features = corr_abs.head(10).index
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df[list(top_features) + ["TARGET"]].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Top Features Associated with Default")
plt.show()
```
